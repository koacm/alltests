Sincos diagrams collection of some graphs using trignometric functions 

Using:
Plotly (python based graphing library)


```python
#please run this before running any other cell
import plotly.express as px 
import pandas as pd 
import plotly.graph_objects as go
import math as m
```

```python
#normal sign wave
a = []
b = []
z = 0
def dtor(deg):
  rad = deg*m.pi/180
  return rad 
for i in range (360):
  try:
    z = dtor(i)
    b.append(i)
    a.append(m.sin(z))
  except:
    pass
print(a,b)
fi = px.line(x = b , y = a)
fi.show()
```

```python
#fancy sine knitting pattern
a = []
b = []
z = 0
for i in range (3600):
  try:
    b.append(i)
    a.append(m.sin(i))
  except:
    pass
print(a,b)
fi = px.scatter(x = a , y = b)
fi.show()
```

```python
# inteference of sin and cos 
a = []
b = []
s = []
c = []
z = 0
def dtor(deg):
  rad = deg*m.pi/180
  return rad 
for i in range (3600):
  try:
    z = dtor(i)
    b.append(i)
    a.append(m.sin(z) - m.cos(z))
    s.append(m.sin(z))
    c.append(m.cos(z))
  except:
    pass
print(a,b)

#fi = px.line(x = b , y = a)
#si = px.line(x = b , y = s)
#co = px.line(x = b , y = c)
#fi.show()

fig = go.Figure()


fig.add_trace(go.Scatter(x=b, y=a,
                    mode='lines',
                    name='sin - cos'))
fig.add_trace(go.Scatter(x=b, y=s,
                    mode='lines',
                    name='sin'))
fig.add_trace(go.Scatter(x=b, y=c,
                    mode='lines',
                    name='cos'))

fig.show()
```

```python
# ripple patter sine wave
a = []
b = []
z = 0
for i in range (2600):
  try:
    b.append(i)
    a.append(m.sin(i))
  except:
    pass
print(a,b)
fi = px.scatter(x = b , y = a)
fi.show()
```

```python
# 3d sin wave 
a = []
b = []
z = 0
def dtor(deg):
  rad = deg*m.pi/180
  return rad 
for i in range (3600):
  try:
    z = dtor(i)
    b.append(i)
    a.append(m.cos(z))
  except:
    pass
print(a,b)
fig = px.scatter_3d( x= a, y= b, z= b)
fig.show()
```
